```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
var='2020'
xls=pd.ExcelFile('Data/Consumers/'+var+'data.xlsx')

Cons_df=xls.parse(var+'data',skiprows=3)
Cons_df.drop([Cons_df.index[0],Cons_df.index[1]],inplace=True)
Cons_df=Cons_df.infer_objects()
Cons_df=Cons_df.drop(Cons_df.tail(8).index)
Cons_df=Cons_df.reset_index(False).drop('index',axis=1)
column_name_mod=['States_UTs', 'Domestic', 'Commercial', 'Industrial_LV', 'Industrial_HV',
'Public_Light', 'Traction', 'Agriculture', 'Public_Water', 'Miscellaneous',
'Total']
Cons_df.columns=column_name_mod
Cons_df=Cons_df.sort_values(by='States_UTs')
Cons_df=Cons_df.reset_index(drop=True)
```

```python
var='2020'
xls=pd.ExcelFile('Data/Consumers/'+var+'data.xlsx')

Cons_df=xls.parse(var+'data',skiprows=3)
Cons_df.drop([Cons_df.index[0]],inplace=True)
Cons_df=Cons_df.infer_objects()
Cons_df=Cons_df.drop(Cons_df.tail(8).index)
Cons_df=Cons_df.reset_index(False).drop('index',axis=1)
column_name_mod=['States_UTs', 'Domestic', 'Commercial', 'Industrial_LV', 'Industrial_HV',
'Public_Light', 'Traction', 'Agriculture', 'Public_Water', 'Miscellaneous',
'Total']
Cons_df.columns=column_name_mod
Cons_df=Cons_df.sort_values(by='States_UTs')
Cons_df=Cons_df.reset_index(drop=True)
```

```python
Cons_df.head()
```

```python
match_phrase='Sub'
match_phrase='Tot'
match_phrase='North'
match_phrase='Eastern'
match_phrase='Western'
match_phrase='Southern'

match_phrases=['Sub','Tot','North','Eastern','Western','Southern']

for match_phrase in match_phrases:
    
    mask=Cons_df['States_UTs'].str.startswith(match_phrase)
    row_to_keep=~mask
    Cons_df=Cons_df[row_to_keep]
    Cons_df=Cons_df.sort_values(by='States_UTs')
    Cons_df=Cons_df.reset_index(drop=True)
Cons_df.columns=Cons_df.columns.str.replace('1','')
Cons_df.columns=Cons_df.columns.str.replace('&','and')
```

```python
plt.figure(figsize=(8, 6))
plt.bar(Cons_df['States_UTs'][:-1],Cons_df['Commercial'][:-1])
# plt.title('Sales Over the Years')
plt.xlabel('States')
plt.ylabel('Consumption (KW)',rotation=90)
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
Cons_df.to_csv('Data/Consumers/processed/consumer_'+var+'.csv')
```
